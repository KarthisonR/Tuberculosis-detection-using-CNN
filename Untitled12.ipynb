```python
## CNN Model


layers = [
    Conv2D(64, (3, 3), activation='relu', input_shape=(224, 224, 1)),#64filter of3*3 to 222,222,64
    MaxPooling2D((2, 2)),#2*2layer to 111,111,64
    Conv2D(128, (3, 3), activation='relu'),#128 filters of3*3 to 109,109,128
    MaxPooling2D((2, 2)),#2*2layer to 54,54
    Conv2D(256, (3, 3), activation='relu'),#256 filters of3*3 to 52,52
    MaxPooling2D((2, 2)),#
    Conv2D(512, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(1024, activation='relu'),
    Dense(512, activation='relu'),
    Dense(2, activation='softmax')
]

# Create a Sequential model and pass the layers list
model_3 = Sequential(layers)

# Compile the model
model_3.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',  # Use 'accuracy' as the loss only if it is appropriate for your use case
    metrics=[Precision(), Recall()]
)
```

```python
model_3.fit(X_train,y_train,epochs=20,validation_data=(X_test,y_test),batch_size=64,verbose= True)
```
